## First Conclusion:

The people who have subscribed and not subscribed have the same distribution. It is necessary to transform nominal and ordinal values ‚Äã‚Äãinto numerical values ‚Äã‚Äãto calculate the correlation between any variable and the target variable.

## Working with Ordinal Values

The target variable is 'has_subscribed'. Verifying the correlation between other variables and the target variable is necessary. Before, It was necessary to change the ordinal variables to numeric; we can do this using OneCodeEncoder from the sci-kit-learn library.

```python
df.head()
```

```python
print_unique_values(df)
```

## The nominal columns will transform using OneHotEncoder:

| Column  | type | transform |
| -------- | -------- | -------- |
| housing | nominal | yes = 1, no = 0 |
| loan |  nominal| yes = 1, no = 0 |
| has_subscribed | nominal | yes = 1, no = 0 |


```python
df.columns
```

```python
from sklearn.preprocessing import OneHotEncoder

columns_to_encode = ['housing', 'loan', 'has_subscribed']

encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_columns = encoder.fit_transform(df[columns_to_encode])
encoded_df = pd.DataFrame(encoded_columns, columns=encoder.get_feature_names_out(columns_to_encode))
df2 = pd.concat([df, encoded_df], axis=1)
df2 = df2.drop(columns=columns_to_encode)
```

```python
df2.head()
```

```python
df2.shape
```

```python
df.shape
```

## The other ordinal variables will be assigned numerical values ‚Äã‚Äãof 0..1, the value will be based on expert judgment.



```python
print_unique_values(df2[['job','marital','education', 'contact', 'poutcome']])
```

```python
cat_job = ['housemaid', 'student',  'unemployed', 'retired', 'blue-collar', 'services', 
           'technician', 'entrepreneur', 'self-employed', 'admin.', 'management' ]

cat_marital = ['single', 'divorced', 'married']

cat_education = ['illiterate', 'basic.4y', 'basic.6y', 'basic.9y', 'high.school', 'professional.course', 'university.degree' ]

cat_contact = ['telephone', 'cellular']

cat_poutcome = ['nonexistent', 'failure', 'success']
```

```python
from sklearn.preprocessing import OrdinalEncoder

datos_ord = df2[['job','marital','education', 'contact', 'poutcome']]

encoder_ord = OrdinalEncoder( categories=[cat_job, cat_marital, cat_education, cat_contact, cat_poutcome])

datos_val = pd.DataFrame(encoder_ord.fit_transform(datos_ord),
                         columns=['job_v','marital_v','education_v', 'contact_v', 'poutcome_v'])

datos_val.head()
```

```python
df2 = pd.concat([df2, datos_val], axis=1)
df2 = df2.drop(columns=['job','marital','education', 'contact', 'poutcome'])
df2.head()
```

```python
df3 = pd.DataFrame()
df3 = df2[ ['age','housing_yes','loan_yes','job_v','marital_v','education_v','contact_v','poutcome_v','has_subscribed_yes'] ]
df3.head()
```

```python
matriz_correlacion = df3.corr()

sns.set(style="white") 
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_correlacion, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)

plt.show()
```

```python
matriz_correlacion['has_subscribed_yes']
```

## Select variables with highest correlation

```python
df_to_model = pd.DataFrame()
df_to_model = df3[ ['age', 'marital_v','education_v','contact_v','poutcome_v','has_subscribed_yes'] ]
df_to_model.head()
```

## Regression Model

```python
X = df_to_model.iloc[:, :-1].values
y = df_to_model.iloc[:, -1].values
```

```python
X[:5,]
```

```python
y[:5,]
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 11)
```

```python
print(f'Size of X_train: {X_train.shape}')
print(f'Size of y_train: {y_train.shape}')
print(f'Size of X_test: {X_test.shape}')
print(f'Size of y_test: {y_test.shape}')
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

cm        = confusion_matrix(y_test, y_pred)
accuracy  = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred) # TP/(TP+FP)
recall    = recall_score(y_test, y_pred)

print(cm)
print(f'Accuracy : {100*accuracy:.2f}%')
print(f'Precision: {100*precision:.2f}%')
print(f'Recall   : {100*recall:.2f}%')
```

```python
sns.set(font_scale=1.2)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")

plt.show()
```

## Conclusion:

The Model is good to predict positive values, but not negative, It is possible because there are more '0's than '1's. We can try balancing the dataset with '1's.

## Balancing the dataset

```python
df_to_model.groupby('has_subscribed_yes')['has_subscribed_yes'].count()
```

```python
33987/4258
```

```python
minority_class = df_to_model[df_to_model['has_subscribed_yes'] == 1]

# How many times you need to duplicates
duplication_factor = 6.5

# Randomly duplicate rows from the minority class
duplicated_minority = minority_class.sample(frac=duplication_factor, replace=True, random_state=42)

# Concatena el DataFrame original con las filas duplicadas
balanced_df = pd.concat([df_to_model, duplicated_minority])

# Muestra el DataFrame resultante
print(balanced_df.head())
```

```python
balanced_df.groupby('has_subscribed_yes')['has_subscribed_yes'].count()
```

```python
X_b = balanced_df.iloc[:, :-1].values
y_b = balanced_df.iloc[:, -1].values

from sklearn.model_selection import train_test_split
X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(X_b, y_b, test_size = 0.25, random_state = 11)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_b = sc.fit_transform(X_train_b)
X_test_b  = sc.transform(X_test_b)

from sklearn.linear_model import LogisticRegression
classifier_b = LogisticRegression(random_state = 0)
classifier_b.fit(X_train_b, y_train_b)

y_pred_b = classifier_b.predict(X_test_b)

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

cm_b        = confusion_matrix(y_test_b, y_pred_b)
accuracy_b  = accuracy_score(y_test_b, y_pred_b)
precision_b = precision_score(y_test_b, y_pred_b) # TP/(TP+FP)
recall_b    = recall_score(y_test_b, y_pred_b)

print(cm)
print(f'Accuracy : {100*accuracy_b:.2f}%')
print(f'Precision: {100*precision_b:.2f}%')
print(f'Recall   : {100*recall_b:.2f}%')

sns.set(font_scale=1.2)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")

plt.show()
```

## Predict the original y_test with the second model

```python
y_pred_f = classifier_b.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

cm_f        = confusion_matrix(y_test, y_pred_f)
accuracy_f  = accuracy_score(y_test, y_pred_f)
precision_f = precision_score(y_test, y_pred_f) # TP/(TP+FP)
recall_f    = recall_score(y_test, y_pred_f)

print(cm)
print(f'Accuracy : {100*accuracy_f:.2f}%')
print(f'Precision: {100*precision_f:.2f}%')
print(f'Recall   : {100*recall_f:.2f}%')

sns.set(font_scale=1.2)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")

plt.show()
```
